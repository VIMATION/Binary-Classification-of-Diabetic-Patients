```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('cleaned__data.csv')
data.head()
```

# Checking Numerical and categorical data

```python
data.dtypes
```

```python
for i in data.describe().columns:
    sns.boxplot(data[i])
    plt.show()
```

```python
for i, row in data.iterrows():
    if row['Pregnancies']>12:
        data=data.drop(i)
    if row['BloodPressure']<40 or row['BloodPressure']>100:
        data=data.drop(i)
    if row['SkinThickness']<15 or row['SkinThickness']>35:
        data=data.drop(i)
    if row['BMI']<15 or row['BMI']>42:
        data=data.drop(i)
    if row['DiabetesPedigreeFunction']>1:
        data=data.drop(i)
    if row['Age']>65:
        data=data.drop(i)
    
```

```python
for i in data.describe().columns:
    sns.distplot(data[i])
    plt.show()
```
