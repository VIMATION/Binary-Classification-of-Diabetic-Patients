## Q.2 
- Plot a Sine Wave with color code with "black" color and style as dashdot, also mark "x" and "y" label and give a suitable title to it

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x),color='black')
plt.xlabel('x',size=30,color='green')
plt.ylabel('y',size=30,color='green')
plt.title('Sine-wave',size=30)
```

## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
x=np.array([2,4,6,8,10])
city=['Delhi','Mumbai','Chennai','Kolkata','Hyderabad']
clothing=[82,78,48,46,39]
Equipment=[48,31,51,29,18]
Accessories=[15,20,14,13,9]
plt.bar(x-.2,clothing,width=0.4,label='clothing')
plt.bar(x+.2,Equipment,width=0.4,label='Equipment')
plt.bar(x+.6,Accessories,width=0.4,label='Accessories')
plt.xticks(x,city,rotation=90)
plt.xlabel('City',size=30)
plt.ylabel('Y',size=30)
plt.legend()
plt.show()
```
